# RISK ANALYSIS

```python
#Calculating Daily Return
def diff(a, b):
    return b - a
data_all['Daily Return'] = data_all.apply(
    lambda x: diff(x['Adj Close'], x['Open']), axis=1)
data_all
```

```python
#calculating mean and standard deviation of the daily return 
#imean and isd is the mean of the daily return

imean=data_all['Daily Return'].iloc[0:989].mean()
print(imean)
isd=data_all['Daily Return'].iloc[0:989].std(axis=0)
print(isd)
```

```python
cmean=data_all['Daily Return'].iloc[2967:3956].mean(axis=0)
print(cmean)
csd=data_all['Daily Return'].iloc[2967:3956].std(axis=0)
print(csd)
```

```python
rmean=data_all['Daily Return'].iloc[1978:2967].mean(axis=0)
print(rmean)
rsd=data_all['Daily Return'].iloc[1987:2967].std(axis=0)
print(rsd)
```

```python
hmean=data_all['Daily Return'].iloc[989:1978].mean(axis=0)
print(hmean)
hsd=data_all['Daily Return'].iloc[989:2967].std(axis=0)
print(hsd)
```

```python
data={'Name':['RELIANCE','HINDUNILVR','CIPLA','INFOSYS'],
      'Expected Return':[rmean,hmean,cmean,imean],  #mean of the daily return of each company
      'Risk':[rsd,hsd,csd,isd],      #standard deviation of the each company
    }
Correlation_data=pd.DataFrame(data)
Correlation_data
```

```python
scatter_plot= px.scatter(Correlation_data, x="Expected Return", y="Risk", color="Name")
scatter_plot.update_traces(marker=dict(size=15,line=dict(width=2,color='Black')))
scatter_plot.show()
```

# Percent Change in Price Analysis

```python
CIPLA_df['percentage_change'] = (CIPLA_df['Close']/CIPLA_df['Close'].shift(1)) -1
INFY_df['percentage_change'] = (INFY_df['Close']/INFY_df['Close'].shift(1))-1
RELIANCE_df['percentage_change'] = (RELIANCE_df['Close']/RELIANCE_df['Close'].shift(1)) - 1
HINDUNILVR_df['percentage_change'] = (HINDUNILVR_df['Close']/HINDUNILVR_df['Close'].shift(1)) -1
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=2, cols=2,subplot_titles=("CIPLA", "INFOSYS", "RELIANCE", "HINDUNILVR"))

fig.add_trace(
    go.Histogram(x=CIPLA_df['percentage_change']),
    row=1, col=1,
    
)

fig.add_trace(
    go.Histogram(x=INFY_df['percentage_change']),
    row=1, col=2
)
fig.add_trace(
    go.Histogram(x=RELIANCE_df['percentage_change']),
    row=2, col=1
)
fig.add_trace(
    go.Histogram(x=HINDUNILVR_df['percentage_change']),
    row=2, col=2
)

fig.update_layout(height=600, width=800, title_text="Price change")
fig.show()
```

It is clear from the graph that the percentage increase in the closing price is highest for infosys . Hence Infosys stock price is the most volatile.

```python
histo=pd.concat([CIPLA_df['percentage_change'],INFY_df['percentage_change'],RELIANCE_df['percentage_change'],HINDUNILVR_df['percentage_change']],axis=1)
histo.columns=['CIPLA(pc)','INFY(pc)','RELIANCE(pc)','HINDUNILVR(pc)']
histo.dropna()

```

# TREND ANALYSIS USING MOVING AVERAGE KEY INDICATOR

OPEN PRICE ANALYSIS

```python
CIPLA_df['SMA20 OPEN'] = CIPLA_df['Open'].rolling(20).mean()
HINDUNILVR_df['SMA20 OPEN'] = HINDUNILVR_df['Open'].rolling(20).mean()
RELIANCE_df['SMA20 OPEN'] = RELIANCE_df['Open'].rolling(20).mean()
INFY_df['SMA20 OPEN'] = INFY_df['Open'].rolling(20).mean()

```

```python
CIPLA_df
```

```python
open_comp=pd.concat([CIPLA_df['Date'],CIPLA_df['SMA20 OPEN'],HINDUNILVR_df['SMA20 OPEN'],RELIANCE_df['SMA20 OPEN'],INFY_df['SMA20 OPEN']],axis=1)
open_comp.columns=['DATE','CIPLA','INFY','RELIANCE','HINDUNILVR']
open_comp=pd.DataFrame(open_comp)
open_comp.dropna()

```

```python
open_comp_plot=px.line(x='DATE',y=['CIPLA','INFY','RELIANCE','HINDUNILVR'],data_frame=open_comp,title="Open Stocks Trend Analysis(SMA) ")
open_comp_plot.show()
```

```python
CIPLA_df['SMA20 CLOSE'] = CIPLA_df['Close'].rolling(20).mean()
HINDUNILVR_df['SMA20 CLOSE'] = HINDUNILVR_df['Close'].rolling(20).mean()
RELIANCE_df['SMA20 CLOSE'] = RELIANCE_df['Close'].rolling(20).mean()
INFY_df['SMA20 CLOSE'] = INFY_df['Close'].rolling(20).mean()
```

```python
close_comp=pd.concat([CIPLA_df['Date'],CIPLA_df['SMA20 CLOSE'],HINDUNILVR_df['SMA20 CLOSE'],RELIANCE_df['SMA20 CLOSE'],INFY_df['SMA20 CLOSE']],axis=1)
close_comp.columns=['DATE','CIPLA','INFY','RELIANCE','HINDUNILVR']
close_comp=pd.DataFrame(close_comp)
close_comp.dropna()
```

```python
close_comp_plot=px.line(x='DATE',y=['CIPLA','INFY','RELIANCE','HINDUNILVR'],data_frame=close_comp,title="Close Stocks Trend Analysis (SMA)")
close_comp_plot.show()
```
