# MOVING AVERAGE

```python
combined_df = pd.read_csv('D:\Study\Project\Stocks-Visualizing-and-Analysis\Datasets\combined.csv',parse_dates=['Date'])
combined_df = combined_df.set_index('Date',drop=False)
```

```python
# Let's go ahead and plot out several moving averages
from pandas import DataFrame
ma_day = [10,20,50]

df1 = DataFrame()
df1['CIPLA']=combined_df['CIPLA']
for ma in ma_day:
    column_name = "MA for %s days" %(str(ma))
    df1[column_name]=pd.DataFrame.rolling(combined_df['CIPLA'],ma).mean()
    
df2 = DataFrame()
df2['HUL']=combined_df['HUL']
for ma in ma_day:
    column_name = "MA for %s days" %(str(ma))
    df2[column_name]=pd.DataFrame.rolling(combined_df['HUL'],ma).mean()
    
df3 = DataFrame()
df3['INFY']=combined_df['INFY']
for ma in ma_day:
    column_name = "MA for %s days" %(str(ma))
    df3[column_name]=pd.DataFrame.rolling(combined_df['INFY'],ma).mean()
    
df4 = DataFrame()
df4['RIL']=combined_df['RIL']
for ma in ma_day:
    column_name = "MA for %s days" %(str(ma))
    df4[column_name]=pd.DataFrame.rolling(combined_df['RIL'],ma).mean()
```

```python
df1[['CIPLA','MA for 10 days','MA for 20 days','MA for 50 days']].plot(figsize=(15,6))
plt.title('Moving average plots for CIPLA Close price')
plt.show()
```

```python
df2[['HUL','MA for 10 days','MA for 20 days','MA for 50 days']].plot(figsize=(15,6))
plt.title('Moving average plots for Hindustan Unilever Limited corp Close price')
plt.show()
```

```python
df3[['INFY','MA for 10 days','MA for 20 days','MA for 50 days']].plot(figsize=(15,6))
plt.title('Moving average plots for INFOSYS renewable energy Close price')
plt.show()
```

```python
df4[['RIL','MA for 10 days','MA for 20 days','MA for 50 days']].plot(figsize=(15,6))
plt.title('Moving average plots for Reliance Industries Limited Close price')
plt.show()
```
